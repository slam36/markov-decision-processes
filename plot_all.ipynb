```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(1)
f = open('gridWorldRewardsVI.txt', 'r')
f1 = f.readlines()

numSteps = []
rewards = []
VItime = []

counter = 0
for line in f1:
    #which number
    line = float(line[:-1])
    
    linetype = counter % 3
    if linetype == 0:
        numSteps.append(line)
    elif linetype == 1:
        rewards.append(line)
    elif linetype == 2:
        VItime.append(line / 1000000)
    counter += 1
        
plt.plot(numSteps)
plt.xlabel('Iterations')
plt.ylabel('Number of Steps')
plt.savefig('gridWorldVInumSteps.png')

plt.figure(2)
plt.plot(rewards)
plt.xlabel('Iterations')
plt.ylabel('Reward')
plt.savefig('gridWorldVIRewards.png')

plt.figure(3)
plt.plot(VItime)
plt.xlabel('Iterations')
plt.ylabel('Time (ms)')
plt.savefig('gridWorldVITime.png')

```

```python
plt.figure(1)
f = open('gridWorldVI.txt', 'r')
f1 = f.readlines()

deltas = []
time = []

counter = 0
for line in f1:
    #which number
    line = float(line[:-1])
    
    linetype = counter % 2
    if linetype == 0:
        deltas.append(line)
    elif linetype == 1:
        time.append(line / 1000000)
    counter += 1
        

plt.plot(deltas)
plt.xlabel('Iteration')
plt.ylabel('delta')
plt.savefig('gridWorldVI_deltas.png')

plt.figure(3)
plt.plot(time)
plt.xlabel('Iteration')
plt.ylabel('Time (ms)')
```
